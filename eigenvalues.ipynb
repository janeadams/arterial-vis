```python
import networkx as nx
from arterialvis import morphology
import numpy.linalg
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
edgelist = morphology.get_edgelist('BG0002.CNG.swc')
```

```python
graph = morphology.generate_graph(edgelist)
```

```python
L = nx.laplacian_matrix(graph)
e = list(numpy.linalg.eigvals(L.A))
e.reverse()
```

```python
e
```

```python
fig = px.scatter(x=list(range(0, len(e))), y=e)
fig.update_layout(template='plotly_white')
fig.update_xaxes(title="k")
fig.update_yaxes(title="Eigenvalue")
fig.show()
```
